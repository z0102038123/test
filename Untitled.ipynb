```python
import pygame
import sys
import random

# 系級：(請填入)
# 姓名：(請填入)
# 學號：(請填入)
# 說明：Problem 2: Interactive Candy Crush Game [cite: 81]
# 本程式使用 pygame 模組實作圖形化介面與連鎖消除邏輯。

# ==========================================
# 1. 遊戲參數與初始化設定
# ==========================================
pygame.init()

# 盤面設定 (使用者可依需求修改長寬與糖果種類)
ROWS = 8
COLS = 6
CANDY_TYPES = 5

# 視窗與繪圖參數
CELL_SIZE = 80  # 每一格的邊長 (像素)
WINDOW_WIDTH = COLS * CELL_SIZE
WINDOW_HEIGHT = ROWS * CELL_SIZE

# 定義糖果顏色字典 (對應數字 0~5)
COLORS = {
    0: (40, 40, 40),      # 0 代表消除後的空位 (背景深灰色)
    1: (255, 89, 94),     # 1 號糖果：紅色
    2: (255, 202, 58),    # 2 號糖果：黃色
    3: (138, 201, 38),    # 3 號糖果：綠色
    4: (25, 130, 196),    # 4 號糖果：藍色
    5: (106, 76, 147)     # 5 號糖果：紫色
}

screen = pygame.display.set_mode((WINDOW_WIDTH, WINDOW_HEIGHT))
pygame.display.set_caption("成大 Python - Candy Crush GUI")
clock = pygame.time.Clock()

# ==========================================
# 2. 遊戲核心邏輯函數
# ==========================================
def create_board():
    """隨機生成初始盤面，確保一開始就充滿糖果"""
    return [[random.randint(1, CANDY_TYPES) for _ in range(COLS)] for _ in range(ROWS)]

def draw_board(board, selected_cell):
    """根據 2D 陣列狀態，將糖果與選取框畫到畫面上"""
    screen.fill((40, 40, 40)) # 塗滿背景
    
    for r in range(ROWS):
        for c in range(COLS):
            candy_val = board[r][c]
            x = c * CELL_SIZE
            y = r * CELL_SIZE
            
            # 畫出對應顏色的糖果圓形
            if candy_val > 0:
                center_pos = (x + CELL_SIZE // 2, y + CELL_SIZE // 2)
                radius = CELL_SIZE // 2 - 8
                pygame.draw.circle(screen, COLORS[candy_val], center_pos, radius)
            
            # 畫出玩家選取糖果的白色提示框
            if selected_cell == (r, c):
                pygame.draw.rect(screen, (255, 255, 255), (x, y, CELL_SIZE, CELL_SIZE), 3)

def find_matches(board):
    """掃描盤面，回傳所有連線(>=3)的糖果座標 Set"""
    matches = set()
    
    # 橫向掃描 (檢查每一列中，是否有連續三個相同)
    for r in range(ROWS):
        for c in range(COLS - 2):
            if board[r][c] != 0 and board[r][c] == board[r][c+1] == board[r][c+2]:
                matches.update([(r, c), (r, c+1), (r, c+2)])
                
    # 縱向掃描 (檢查每一行中，是否有連續三個相同)
    for c in range(COLS):
        for r in range(ROWS - 2):
            if board[r][c] != 0 and board[r][c] == board[r+1][c] == board[r+2][c]:
                matches.update([(r, c), (r+1, c), (r+2, c)])
                
    return matches

def drop_and_refill(board, matches):
    """執行消除，讓糖果掉落，並隨機生成新糖果從頂部補滿"""
    # 1. 消除：將有連線的位置變成 0
    for r, c in matches:
        board[r][c] = 0
        
    # 2. 掉落與補齊：針對每一個直行 (column) 獨立處理
    for c in range(COLS):
        # 收集這一直行中「沒有被消除」的糖果 (從上到下)
        column_candies = [board[r][c] for r in range(ROWS) if board[r][c] != 0]
        
        # 計算空缺數量
        missing = ROWS - len(column_candies)
        
        # 隨機生成新糖果來填補頂部空位
        new_candies = [random.randint(1, CANDY_TYPES) for _ in range(missing)]
        
        # 組合新糖果與舊糖果
        new_column = new_candies + column_candies
        
        # 更新該直行的狀態
        for r in range(ROWS):
            board[r][c] = new_column[r]

# ==========================================
# 3. 遊戲主迴圈
# ==========================================
board = create_board()
selected = None  # 記錄玩家點擊的第一顆糖果座標 (row, col)

# 遊戲開始前，先進行一次初始盤面的消除 (避免一開局就有連線)
initial_matches = find_matches(board)
while initial_matches:
    drop_and_refill(board, initial_matches)
    initial_matches = find_matches(board)

running = True
while running:
    # A. 處理玩家操作事件
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        
        # 偵測滑鼠點擊
        if event.type == pygame.MOUSEBUTTONDOWN:
            mx, my = event.pos
            
            # 將像素座標轉換回二維陣列的 [row][col] 索引
            clicked_col = mx // CELL_SIZE
            clicked_row = my // CELL_SIZE
            
            if selected is None:
                # 狀態一：記錄第一次點擊的糖果
                selected = (clicked_row, clicked_col)
            else:
                # 狀態二：處理兩顆糖果的交換邏輯
                r1, c1 = selected
                r2, c2 = clicked_row, clicked_col
                
                # 檢查兩顆糖果是否相鄰 (距離為 1)
                if abs(r1 - r2) + abs(c1 - c2) == 1:
                    # 在 2D 陣列中互換兩個數字的值
                    board[r1][c1], board[r2][c2] = board[r2][c2], board[r1][c1]
                    
                    # 檢查交換後是否產生連線
                    matches = find_matches(board)
                    
                    if not matches:
                        # 無效交換：換回來
                        board[r1][c1], board[r2][c2] = board[r2][c2], board[r1][c1]
                        print("無效交換：沒有形成連線！")
                    else:
                        print(f"成功交換: ({r1}, {c1}) 與 ({r2}, {c2})")
                        
                        # 處理連鎖消除與掉落
                        while matches:
                            drop_and_refill(board, matches)
                            
                            # 更新畫面並暫停，製造掉落的視覺動畫效果
                            draw_board(board, None)
                            pygame.display.flip()
                            pygame.time.delay(300) 
                            
                            # 再次掃描是否有新的連線
                            matches = find_matches(board)
                else:
                    print("交換失敗：必須選擇上下左右相鄰的糖果喔！")
                
                # 重置選取狀態
                selected = None

    # B. 繪製並更新畫面
    draw_board(board, selected)
    pygame.display.flip()
    clock.tick(60)

# 安全關閉遊戲
pygame.quit()
sys.exit()
```
